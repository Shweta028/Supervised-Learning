```python
#!usr/bin/python

from sklearn import tree
from sklearn.datasets import load_iris
import numpy as np

#loading data from iris_database and printing attributes

iris=load_iris()
print dir(iris)

#print flower names

flw_names=iris.target_names
print (flw_names)

#print features of flowers

flw_features=iris.feature_names
print (flw_features)

#print target values
print (iris.target)

#loading value of features

flw_feature_data=iris.data

#Dividing dataset into training data and test data
#IRIs contains 150 sample data rows: 50 of each category
#Let us take 1 sample from each category for testing and rest for training

#Creating training data 
x=[0,50,100]

train_data=np.delete(flw_feature_data,x,axis=0)  #delete fucntion from numpy deletes selected data from given list

test_data=flw_feature_data[x]

#Creating test data

flw_real_target=iris.target
train_target=np.delete(flw_real_target,x)   #removing same data from target

print train_data.size
print train_target.size
print test_data.size

clf=tree.DecisionTreeClassifier()
trained=clf.fit(train_data,train_target)
out=trained.predict(test_data)
print ("The respective categories are ")
print out

#Taking user value to predict
ch=input("Press 1 for User Input")
inputfeature=[]
if ch==1:
    print ("Enter 4 features: ")
    for i in range(4):
        a=input()
        inputfeature.append(a)
    
    output=trained.predict([inputfeature])

    if output==[0]:flower="Setosa"
    elif output==[1]: flower="Versicolor"
    else: flower="Virginica"
    print ("The given features are of " +flower+ " flower.")
```
