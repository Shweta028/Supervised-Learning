```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor as knn
from sklearn.model_selection import train_test_split


#loading breast cancer dta
info = load_breast_cancer()

#For understanding a new data set, we can print its complete description by DESCR command
print info.DESCR

#Performing automatic division in training and testing data by using train test split
train_data,test_data,train_target,test_target=train_test_split(info.data, info.target, test_size=0.01)

#calling knn with no. of neighbors=5
clf=knn(n_neighbors=5)

#training the machine
trained=clf.fit(train_data,train_target)

#predicting target of test data by trained machine
predicted=trained.predict(test_data)

#Since breast cancer data has too many attributes, we can perform feature reduction 
#printing test_data
print (test_data)

#print output of prediction
print (predicted)


```
